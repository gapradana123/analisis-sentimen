```python
from google_play_scraper import reviews, Sort
import pandas as pd
```

# Scrapping

```python
# Mengambil semua ulasan dari aplikasi 'mytelkomsel di Google Play Store.
scrapreview, continuation_token = reviews(
    'com.telkomsel.telkomselcm', # ID aplikasi 'mytelkomsel'
    lang='id',                   # Ulasan dengan bahasa indonesia
    country='id',                # Negara Indonesia
    sort=Sort.NEWEST,            # Urutan ulasan mulai dari yang paling baru
    count=100000                  # Ambil 100000 ulasan terbaru
)
```

```python
# Membuat dataframe dari hasil scrapping
app_reviews_df = pd.DataFrame(scrapreview)
# Menampilkan informasi tentang DataFrame app_reviews_df
app_reviews_df.info()
```

```python
# Memilih kolom yang digunakan
app_reviews_df = app_reviews_df[['content', 'score']]
# Membuat DataFrame baru (clean_df) dengan menghapus baris yang memiliki nilai yang hilang (NaN) dari app_reviews_df
clean_df = app_reviews_df.dropna()
# Menghapus baris duplikat dari DataFrame clean_df
clean_df = clean_df.drop_duplicates()
# Reset index
clean_df = clean_df.reset_index(drop=True)
# Menampilkan informasi tentang DataFrame clean_df setelah membersihkan nilai yang hilang (NaN) dan duplikat
clean_df.info()
```

```python
pd.set_option('display.max_colwidth', 3000)
# Menampilkan dataframe
clean_df
```

```python
# Menyimpan hasil scrapping dalam bentuk csv
clean_df.to_csv('ulasan_aplikasi_mytelkomsel.csv', index=False)
```
